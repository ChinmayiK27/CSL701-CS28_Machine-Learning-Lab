Task 7: Generate summary statistics using df.describe(); record mean, minimum and
maximum values of Age and Fare.

```python
summary_stats = df.describe()
display(summary_stats)

# Record mean, minimum and maximum values of Age and Fare
age_mean = summary_stats.loc['mean', 'Age']
age_min = summary_stats.loc['min', 'Age']
age_max = summary_stats.loc['max', 'Age']

fare_mean = summary_stats.loc['mean', 'Fare']
fare_min = summary_stats.loc['min', 'Fare']
fare_max = summary_stats.loc['max', 'Fare']

print(f"\nAge - Mean: {age_mean:.2f}, Min: {age_min:.2f}, Max: {age_max:.2f}")
print(f"Fare - Mean: {fare_mean:.2f}, Min: {fare_min:.2f}, Max: {fare_max:.2f}")
```

Task 9: Fill missing Age values with the median.

Task 10: Filter surviving passengers (Survived == 1).

```python
survived_passengers = df[df['Survived'] == 1]
display(survived_passengers.head())
```

Task 11: Convert Fare column to a NumPy array.

```python
fare_array = df['Fare'].to_numpy()
print(f"Type of fare_array: {type(fare_array)}")
display(fare_array[:5]) # Display first 5 elements to verify
```

Task 12: Calculate maximum and average Fare using np.max() and np.mean().


```python
max_fare = np.max(fare_array)
mean_fare = np.mean(fare_array)

print(f"Maximum Fare: {max_fare:.2f}")
print(f"Average Fare: {mean_fare:.2f}")
```

Task 13: Plot histogram of Age.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Task 14: Plot bar chart of Survival Count.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = Not Survived, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Task 15: Plot scatter graph of Age vs Fare.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=df.dropna(subset=['Age', 'Fare']))
plt.title('Scatter Plot of Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
